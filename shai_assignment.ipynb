### Descriptive Statistics

```python
# generating descriptive statistics of the dataset
data.describe()
```

```python
# calculating the mean, median, mode , min , max, and standard deviation of salary 

salary= data['TotalPay']
mean_salary = np.mean(salary)
print(f" the mean salary is : {mean_salary}")

# calculating the median of salary
median= salary.median()
print(f" the median of salary is : {median}")

# calculating the mode 
mode = salary.mode()
print(f" the mode is : {mode[0]}")

# calculating the min and max salary
min_salary = min(salary)
print(f" The minimum salary is : {min_salary}")
max_salary = max(salary)
print(f" The maximum salary is :{max_salary}")

# determining the salary range 
salary_range = max_salary - min_salary
print (f" the salary range is : {salary_range}")

# caculating te standard deviation
std_deviation = salary.std()
print(f" the standard deviation is {std_deviation}")
```

### Handling Missing Data

Handling the missing data using mean() function is a common approach , espacially for numerical columns ,so I choose this method as it provides a simple and reasonable estimate of the missing values.
I also dropped the Notes and Status columns as they don't contain any values so dropping them is the suitable way for handling the missing data. 

```python
# handling the missing data 
data['BasePay'].fillna(data['BasePay'].mean(), inplace = True)
data['OvertimePay'].fillna(data['OvertimePay'].mean(), inplace = True)
data['OtherPay'].fillna(data['OtherPay'].mean(), inplace = True)
data['Benefits'].fillna(data['Benefits'].mean(), inplace = True)

data.drop(columns=["Notes","Status"], inplace = True)

data.isnull().sum()
```

### Data Visualization

```python
# visualizing the distribtion of salaries using Histogram
plt.hist(data['TotalPay'], bins=20, color ='blue', edgecolor='black')
plt.title("distribution of salaries")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
data['JobTitle']=data['JobTitle'].str.lower()
grouped = data.groupby(['JobTitle'])['Id'].agg('count')
plt.pie(grouped, labels = grouped)
plt.title("the proportion of employees in different departments")
plt.show()
```

### Grouped Analysis

```python
# Grouping the data by Job Title column
grouped_data = data.groupby(['JobTitle'])['BasePay','OvertimePay','OtherPay',
                                                   'Benefits','TotalPay','TotalPayBenefits'].agg(
                                                  ['mean', 'median', 'count', 'sum','std'])
grouped_data
```

### Simple Correlation Analysis

```python
# correlation between salary and the other columns
corr1 = data['TotalPay'].corr(data['Year']) # 0.03
corr2 = data['TotalPay'].corr(data['OvertimePay']) # 0.50
corr3 = data['TotalPay'].corr(data['BasePay']) #0.95
corr4 = data['TotalPay'].corr(data['OtherPay']) # 0.4
corr5 = data['TotalPay'].corr(data['TotalPayBenefits']) # 0.97
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='TotalPay', y='TotalPayBenefits', data= data)
plt.title("The relationship between Salary and TotalPayBenefits")
plt.xlabel("Salary")
plt.ylabel("TotalPayBenefits")
plt.show()
```

```python
data = {'X-axis':data['TotalPay'], 'TotalPayBenefits': data['TotalPayBenefits'], 'BasePay': data['BasePay'],
        'OvertimePay':  data['OvertimePay'], 'OtherPay':  data['OtherPay']}
df = pd.DataFrame(data)

g = sns.FacetGrid(df.melt(id_vars='X-axis'), col='variable', col_wrap=2, sharey=False)
g.map(sns.scatterplot, 'X-axis', 'value')

# Add labels
g.set_axis_labels('TotalPay', 'Y-axis')

# Show the plot
plt.show()
```
